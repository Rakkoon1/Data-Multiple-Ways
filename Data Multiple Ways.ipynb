## Choose one continuous variable and one categorical variable, and plot them six different ways:

My one continuous variable is wine compared to the categorical variable 'continent' to examine the consumption habits for wine based on each continent. By using different types of factorplots we can compare the visibility of the data and which is best suited for analysis.

```python
sns.factorplot(x='continent', y='wine', data=drinks, kind='bar');
plt.title('Plot 1: Wine Consumption by Continent')
plt.show();
```

```python
sns.factorplot(x='continent', y='wine', data=drinks, kind='box');
plt.title('Plot 2: Wine Consumption by Continent')
plt.show();
```

```python
sns.factorplot(x='continent', y='wine', data=drinks);
plt.title('Plot 3: Wine Consumption by Continent')
plt.show();
```

```python
sns.factorplot(x='continent', y='wine', data=drinks, kind='swarm');
plt.title('Plot 4: Wine Consumption by Continent')
plt.show();
```

```python
sns.factorplot(x='continent', y='wine', data=drinks, kind='violin');
plt.title('Plot 5: Wine Consumption by Continent')
plt.show();
```

```python
sns.factorplot(x='continent', y='wine', data=drinks, kind='strip');
plt.title('Plot 6: Wine Consumption by Continent')
plt.show();
```

The boxplot provides the best visibility for analysis with clear points for outliers and distribution. The swarm is quite clear in displaying where the data points cluster and the location of outliers. The strip plot is less informative about clusters but still provides clear insight to outliers. The bar plot is very useful for clearly displaying the differences between continents but it does not provide insight on outliers. The point plot is straight to the point but similar to the bar plot in providing less insight on data points. The violin plot is my least preferred for this data because it displays values below zero and I imagine it may be difficult to explain to others who may question why people are consuming less than none.

## Probability Distributions

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
sns.boxplot(x='continent', y='beer', data=drinks)

plt.subplot(2,2,2)
sns.boxplot(x='continent', y='wine', data=drinks)

plt.subplot(2,2,3)
sns.boxplot(x='continent', y='spirits', data=drinks)

plt.subplot(2,2,4)
sns.boxplot(x='continent', y='spirit_litres', data=drinks)
plt.show();
```

```python
from scipy.stats import norm

sns.distplot(drinks['beer'], bins=10, fit=norm);
```

```python
sns.distplot(drinks['spirits'], bins=10, fit=norm);
```

```python
sns.distplot(drinks['wine'], bins=10, fit=norm);
```

```python
sns.distplot(drinks['spirit_litres'], bins=10, fit=norm);
```

```python
plt.scatter(x=drinks['spirit_litres'], y=drinks['spirits'], label='Spirits', color='purple')
plt.title('Spirits and Spirit Litres')
plt.show()
```

```python
plt.scatter(x=drinks['wine'], y=drinks['beer'], label='Spirits', color='blue')
plt.title('Wine and Beer')
plt.show()
```

```python
print(drinks['beer'].sum())
print(drinks['wine'].sum())
print(drinks['spirits'].sum())
print(drinks['spirit_litres'].sum())
```

For each of the continuous variables the data is skewed to the right. The density curve reflects how there are many countries which either do not have as much access to a variety of alcohol compared to others or simply prefer consuming one type more than the other. In this sample we can see that wine has the highest density spike that peaks at zero. Because of the other countries which consume high amounts of wine, the normalization curve peaks roughly between 0 and 100. We can also see that beer has the lowest density spike at roughly .005. Based on the visuals from the sample data, I hypothesised that beer has the highest consumption across all the continents. Finally, to check my hypothesis I just simply used .sum() for each and found it to be true by beating spirits by nearly 5,000.
